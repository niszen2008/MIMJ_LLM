```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

model_name = "mistralai/Mistral-7B-Instruct-v0.2"

tokenizer = AutoTokenizer.from_pretrained(model_name)

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16,
    device_map="auto"
)

print("AI Brain Loaded Successfully 🧠")
```

```python
def ask_ai(question):
    inputs = tokenizer(question, return_tensors="pt").to(model.device)
    output = model.generate(**inputs, max_new_tokens=150)
    return tokenizer.decode(output[0], skip_special_tokens=True)

ask_ai("I want to learn Hindi. What should I do?")
```

```python
import gradio as gr

def chat(message):
    inputs = tokenizer(message, return_tensors="pt").to(model.device)
    output = model.generate(**inputs, max_new_tokens=200)
    return tokenizer.decode(output[0], skip_special_tokens=True)

gr.Interface(
    fn=chat,
    inputs="text",
    outputs="text",
    title="My Personal LLM 🤖",
    description="Ask anything!"
).launch()
```

```python
import pandas as pd

file_path = "/content/quran-data.xlsx"  # change filename
df = pd.read_excel(file_path)

df.head()
```

```python
texts = []

for index, row in df.iterrows():
    row_text = ", ".join([f"{col}: {row[col]}" for col in df.columns])
    texts.append(row_text)

texts[:2]
```

```python
from langchain_core.documents import Document
documents = [Document(page_content=text) for text in texts]
len(documents)
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

db = FAISS.from_documents(documents, embeddings)

print("Excel data stored in AI memory ✅")
```

```python
query = "What are the main details in this excel?"
results = db.similarity_search(query, k=3)

for r in results:
    print(r.page_content)
    print("-----")
```

```python
def chat_with_excel(question):
    # Step A: Search Excel memory
    docs = db.similarity_search(question, k=3)

    # Step B: Combine Excel rows into context
    context = "\n".join([d.page_content for d in docs])

    # Step C: Tell AI to use ONLY Excel data
    prompt = f"""
    You are an assistant.
    Answer ONLY using the Excel data below.
    If answer is not found, say "Not available in Excel".

    Excel Data:
    {context}

    Question:
    {question}
    """

    # Step D: Ask the AI model
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    output = model.generate(**inputs, max_new_tokens=200)

    return tokenizer.decode(output[0], skip_special_tokens=True)
```

```python
chat_with_excel("Summarize this Excel file")
```

```python
import gradio as gr

gr.Interface(
    fn=chat_with_excel,
    inputs="text",
    outputs="text",
    title="Excel AI Assistant 📊",
    description="Ask questions from your Excel file"
).launch()
```

```python
def contains_search(question, texts, top_k=5):
    question_lower = question.lower()
    matches = []

    for text in texts:
        if question_lower in text.lower():
            matches.append(text)

    return matches[:top_k]
```

```python
def hybrid_search(question, top_k=5):
    # A) Contains search
    contains_results = contains_search(question, texts, top_k)

    # B) Vector (meaning) search
    vector_docs = db.similarity_search(question, k=top_k)
    vector_results = [d.page_content for d in vector_docs]

    # C) Merge + remove duplicates
    combined = list(dict.fromkeys(contains_results + vector_results))

    return combined[:top_k]
```

```python
def chat_with_excel(question):
    results = hybrid_search(question)

    if not results:
        return "Not available in Excel"

    context = "\n".join(results)

    prompt = f"""
    You are an assistant.
    Answer ONLY using the Excel data below.
    Do NOT guess.

    Excel Data:
    {context}

    Question:
    {question}
    """

    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    output = model.generate(**inputs, max_new_tokens=200)

    return tokenizer.decode(output[0], skip_special_tokens=True)
```

```python
import gradio as gr

gr.Interface(
    fn=chat_with_excel,
    inputs="text",
    outputs="text",
    title="Excel Smart AI Assistant 📊",
    description="Keyword + Meaning based search"
).launch()
```
